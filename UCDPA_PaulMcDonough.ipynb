```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Define Function for getting data set info#
def GetDatasetOverview(dataset_df):
        return print("Data set Info"),print(dataset_df.info()),print(),print("Dataset head"),print(dataset_df.head())
```

```python
#Upload F1 races data#
f1_races_df = pd.read_csv('races.csv')
```

```python
#Get overview of F1 races dataset#
GetDatasetOverview(f1_races_df)
```

```python
#Upload F1 Circuits data & get overview of dataset#
f1_circuits_df=pd.read_csv('circuits.csv')
GetDatasetOverview(f1_circuits_df)
```

```python
#Merge F1 Races and Circuits data#
f1 = pd.merge(f1_races_df, f1_circuits_df,on='circuitId')
```

```python
#Datset Overview#
GetDatasetOverview(f1)
```

```python
#Rename column to allow merge with Human Freedom Index Dataset#
f1.rename(columns={'country':'countries'}, inplace=True)
```

```python
#Add aditional column confirming sport as F1#
f1['Sport']='F1'
```

```python
#Filter Dataset to relevant information#
f1_host=f1[['year','countries', 'Sport']]
```

```python
#Get new datast overview#
GetDatasetOverview(f1_host)
```

```python
#Iterrows function to view all countries -> Required for data cleaning to ensure country naming conventions are consistent with HFI Dataset#
for row in f1_host.iterrows():
    print(row[1]['countries'])
```

```python
#Import Human Freedom Index Dataset#
freedom_index_df=pd.read_csv('hfi_cc_2021.csv')
```

```python
#Get new datast overview#
GetDatasetOverview(freedom_index_df)
```

```python
for col_name in freedom_index_df.columns: 
    print(col_name)
```

```python
#Check country naming conventions in HFI dataset#
for row in freedom_index_df.iterrows():
    print(row[1]['countries'])
```

```python
#Rename countries in HFI Dataset to align with naming conventions in F1 dataset#
freedom_index_df['countries']=freedom_index_df['countries'].replace(['United States'],['USA'])
freedom_index_df['countries']=freedom_index_df['countries'].replace(['Russian Federation'],['Russia'])
freedom_index_df['countries']=freedom_index_df['countries'].replace(['Korea, Rep.'],['Korea'])
freedom_index_df['countries']=freedom_index_df['countries'].replace(['United Kingdom'],['UK'])
freedom_index_df['countries']=freedom_index_df['countries'].replace(['United Arab Emirates'],['UAE'])
```

```python
#Filter Freedom Index Dataset to relevant values#
freedom_rank_df=freedom_index_df[['year', 'countries', 'hf_score','hf_rank','region']]
print(freedom_rank_df.head)
```

```python
freedom_rank_df.dtypes
```

```python
#Function for joining datasets with HFI dataset - relevant should we want to make comparisons with other sports#
def joinfreedom(sport_df):
        return pd.merge(sport_df, freedom_rank_df, on=['year','countries'], how='outer')
```

```python
#Pass F1 dataset through merging function#
F1 = joinfreedom(f1_host)
```

```python
print(F1.head())
```

```python
F1[F1.isnull().any(axis=1)]
```

```python
#Filter values where 'Sport' is null as they are not F1 host countries#
#Filter between 2008 and 2019, as this is the years the HFI dataset tracks#
F1_filtered=F1[F1['Sport'].notnull()]
F1_filtered2=F1_filtered[F1_filtered['year'].between(2008,2019)]
print(F1_filtered2.info())
print(F1_filtered2.tail(5))
```

```python
#Checkfor any null values -> Monaco is not tracked in the HFI Index and will need to be excluded#
F1_filtered2[F1_filtered2.isnull().any(axis=1)]
```

```python
#Drop null values -> Monaco#
F1_filtered3=F1_filtered2.dropna()
GetDatasetOverview(F1_filtered3)
```

```python
#Data Analysis#
```

```python
#HFI interquartile values for all included countries between 2008 and 2019#
freedom_index_df1=freedom_index_df[freedom_index_df['hf_score'].notnull()]
freedom_index_df2=freedom_index_df1[freedom_index_df1['year'].between(2008,2019)]
```

```python
print('F1 Host Median 50% is', np.percentile(F1_filtered3.hf_score, 50))
```

```python
#Data Visualisation#
```

```python
#Boxplot to visualise the spread of F1 Host countries HFI scores over time#
ax1 = sns.boxplot(x="year", y="hf_score", data=F1_filtered3, palette='flare')
ax1.set_xlabel("Year", fontsize=10)
ax1.set_ylabel('Human Freedom Score', fontsize=10)
ax1.set_title('F1 Host Countries HFI Scores', fontsize=15)
plt.savefig('ax1')
```

```python
#Visualising the Interquartile range of F1 Host countries HFI scores#
ax2 = sns.lineplot(x="year", y="hf_score", estimator = np.median, 
                  data=F1_filtered3,ci=None)
bounds = F1_filtered3.groupby('year')['hf_score'].quantile((0.25,0.75)).unstack()
ax2.fill_between(x=bounds.index,y1=bounds.iloc[:,0],y2=bounds.iloc[:,1],alpha=0.3)
ax2.set_xlabel("Year", fontsize=10)
ax2.set_ylabel('Human Freedom Score', fontsize=10)
ax2.set_title('F1 Host Countries HFI Interquartile Range', fontsize=15)
plt.ylim(6,9)
plt.savefig('ax2')
```

```python
#Visualising the interquartile range of the HFI index for all countries over the same period#
ax3 = sns.lineplot(x="year", y="hf_score", estimator = np.median, 
                  data=freedom_index_df2,ci=None)
bounds = freedom_index_df2.groupby('year')['hf_score'].quantile((0.25,0.75)).unstack()
ax3.fill_between(x=bounds.index,y1=bounds.iloc[:,0],y2=bounds.iloc[:,1],alpha=0.3)
ax3.set_xlabel("Year", fontsize=10)
ax3.set_ylabel('Human Freedom Score', fontsize=10)
ax3.set_title('Human Freedom Index Interquartile Range', fontsize=15)
plt.ylim(6,9)
plt.savefig('ax3')
```
